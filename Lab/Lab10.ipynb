```python
class student():
    def __init__(self):
        self.dictnry={}
    def details(self,name,roll,email,m1):
        temp={"name":name,
              "roll":roll,
              "email":email,
              "mark":m1
              }
        self.dictnry[roll]=temp
    def search(self,roll):
        if roll in self.dictnry:
            print(self.dictnry)
    def delete(self,roll):
        if roll in self.dictnry:
            del self.dictnry[roll]
    def edit(self,roll,ed,val):
        if roll in self.dictnry:
            self.dictnry[roll][ed]=val
            print("value changed")
    def view(self):
        print(self.dictnry)
pd=student()
while True:
    print("1.create student")
    print("2.search student")
    print("3. delete student")
    print("4.edit student")
    print("5.view student")
    print("6.Exit")
    choice= int (input("Enter your choice:"))
    if choice==1:
        name=input("Enter the name of student:")
        roll=input("Enter the RollNo:")
        email=input("Enter the Email ID:")
        m1=input("Enter the mark:")
        pd.details(name,roll,email,m1)
        print("Added Sucessfully")
    if choice==2:
        roll=input("Enter the RollNo:")
        pd.search(roll)
    if choice==3:
        roll=input("Enter the RollNo:")
        pd.delete(roll)
        print("Value Deleted Sucessfully")
    if choice==4:
        roll=input("Enter the RollNo:")
        ed=input("Enter the field to edited")
        val=input("Enter the value you to edited")
        pd.edit(roll,ed,val)
        print("Value edited sucessfully:")
    if choice==5:
        pd.view()
    if choice==6:
        break
```
