```python
import random
item_list=["Apple","Banana","mango","kiwi","orange","grapes","Guava","cherry","papaya","coco"]
Quantity = ["5","6","2","4","6","4","6","10","1","2"]
price=["250","50","100","90","80","230","60","110","30","55"]
def Gitem(item_list,Quantity,price):
    file=open("Mix & Match Calculator.csv", "w+")
    print("Item\t\tQuantity\tPrice\t\ttotal")
    file.write("Item:\t\tQuantity:\tPrice:\t Total\n")
    for a in range(0,100): 
        item= random.randint(0,9)
        qnty=random.randint(0,9)
        prc=random.randint(0,9)
        total=int(Quantity[qnty])*int(price[prc])
        print(item_list[item],"\t\t",str(Quantity[qnty]),"\t\t",str(price[prc]),"\t\t",str(total))
        file.write(item_list[item]+"\t\t"+Quantity[qnty]+"\t\t\t"+price[prc]+"\t\t"+str(total)+"\n")
Gitem(item_list,Quantity,price)
```
